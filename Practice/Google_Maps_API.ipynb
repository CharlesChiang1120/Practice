```python
import requests
import googlemaps
import pandas as pd
```

```python
cities = ['臺北市','新北市','臺中市','臺南市','高雄市','基隆市','嘉義市','台南市','新竹市','高雄市']
```

```python
geocode_result = gmaps.geocode('中和區')
geocode_result
```

```python
loc = geocode_result[0]['geometry']['location']
loc
```

```python
ids = []
keyword = '石頭火鍋'

for city in cities:
    geocode_result = gmaps.geocode(city)
    loc = geocode_result[0]['geometry']['location']
    print('以'+city + "為中心半徑25000公尺的{}的數量".format(keyword)+'' + str(len(gmaps.places_nearby(keyword = keyword,location=loc,\
                                                                         radius = 25000)['results'])))
    for place in gmaps.places_nearby(keyword = keyword,location=loc,radius = 25000)['results']:
        ids.append(place['place_id'])
```

```python
stors_info = []
#set去重複 再用List還原回來
ids = list(set(ids))
for id in ids:
    stors_info.append(gmaps.place(place_id = id ,language = 'zh-TW')['result'])
```

```python
output = pd.DataFrame(stors_info,)
output
```

```python
output['lat'] = output['geometry'].map(lambda x : x['location']['lat'])
```

```python
output['lng'] = output['geometry'].map(lambda x : x['location']['lng'])
```
